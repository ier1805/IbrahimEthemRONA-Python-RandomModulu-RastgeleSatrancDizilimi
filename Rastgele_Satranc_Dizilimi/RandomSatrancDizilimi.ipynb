### **PRINT ILE OLUSMUS RASTGELE SATRANC** 

```python
import random

tahta = [["8", " ", " ", " ", " ", " ", " ", " ", " ", "8"], ["7", " ", " ", " ", " ", " ", " ", " ", " ", "7"], ["6", " ", " ", " ", " ", " ", " ", " ", " ", "6"], ["5", " ", " ", " ", " ", " ", " ", " ", " ", "5"], ["4", " ", " ", " ", " ", " ", " ", " ", " ", "4"], ["3", " ", " ", " ", " ", " ", " ", " ", " ", "6"], ["2", " ", " ", " ", " ", " ", " ", " ", " ", "2"], ["1", " ", " ", " ", " ", " ", " ", " ", " ", "1"]]

taslar = ["♖","♖","♜","♜","♔","♚","♕","♛","♗","♗","♝","♝","♘","♘","♞","♞","♙","♙","♙","♙","♙","♙","♙","♙","♟","♟","♟","♟","♟","♟","♟","♟"]

indis = 0
while (indis < 32):
    satir = random.randint(0,7)
    sutun = random.randint(0,9)
    if (tahta[satir][sutun] == " "):
        tahta[satir][sutun] = taslar[indis]
        indis = indis + 1

print("    A   B   C   D   E   F   G   H   ")

for i1 in range(0,8):
    if (i1 == 0):
        print("  ┌───┬───┬───┬───┬───┬───┬───┬───┐")
    metin = tahta[i1][0] + " │ "
    for i2 in range(0,9):
        if (i2 == 8):
            metin = metin + tahta[i1][i2 + 1]
        else:
            metin = metin + tahta[i1][i2 + 1] + " │ "
    print(metin)
    if (i1 != 7):
        print("  ├───┼───┼───┼───┼───┼───┼───┼───┤")
    metin = " │ "
    if (i1 == 7):
        print("  └───┴───┴───┴───┴───┴───┴───┴───┘")

print("    A   B   C   D   E   F   G   H   ")
```
